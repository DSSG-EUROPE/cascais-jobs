#2. Education attainment by age

```python
# See number of apps submitted for each education level by age. Check that there aren't really young people getting doctorates
#Create age v. education DataFrame and sort on education level
edu_age = DataFrame(applications.groupby(['ute_idade', 'education']).ute_id.count())
edu_age = edu_age.reset_index().pivot('education', 'ute_idade', 'ute_id')
edu_order = ['cannot read/write', 'grade school', 
            '04', '06', '09', '11', '12', 'post-secondary', 'licensed',
            'bachelor', 'masters', 'doctoral']
edu_age = edu_age.reindex(index = edu_order[::-1])

#plot number of apps, up to the first 30 ages
fig, ax = plt.subplots(figsize = (16,5))
sns.heatmap(edu_age.iloc[:, :30], annot=True, fmt='g', ax=ax)
```

#5. Discrepencies between application date and anomes

```python
#Check how often year_month is different from anomes

#make datetime objects of anomes and candidatura_data
app_dates = applications[['anomes', 'candidatura_data']]
app_dates['trunc_date'] = app_dates['candidatura_data'].apply(lambda x: pd.to_datetime(x[0:4]+x[5:7], format='%Y%m', errors='coerce'))
app_dates['year_month'] = app_dates['anomes'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m', errors='coerce'))
app_dates['date_mismatch'] = app_dates['year_month']!=app_dates['trunc_date']
app_dates.head()

#select only those with mismatches
mismatches = app_dates[app_dates['date_mismatch']==True]
mismatches['difference'] = mismatches['year_month']-mismatches['trunc_date']
```

```python
#plot mismatches
#plot number of apps, up to the first 30 ages
fig, ax = plt.subplots(figsize = (16,5))
mismatches.difference.value_counts().sort_index().plot(kind='bar', ax=ax)
```
